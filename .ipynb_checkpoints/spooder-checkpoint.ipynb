```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/Car_Purchasing_Data.csv")
df.head()
```

```python
df["Age"] = df["Age"].astype(int)   # WHY ARE THE AGESS IN FLOAT LOL
```

```python
sns.pairplot(df)
```

```python
plt.scatter(df["Age"], df["Credit Card Debt"])
```

```python
# less drop the name and email and country cos they prob dont affect lool
df.columns
```

```python
X = df.drop(['Customer Name', 'Customer e-mail', 'Country'], axis=1)
X.head()
```

```python
y = np.array(df["Car Purchase Amount"])
y[:4]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled[:4]
```

```python
y = y.reshape(-1,1)  # have to do this cos scaler expects a > 1D data
y_scaled = scaler.fit_transform(y)
y_scaled[:4]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2)
```

```python
X_train.shape
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense, Input
```

```python
# Creating the neural nett
model = Sequential()
model.add(Input(shape=(6,)))
model.add(Dense(25, activation = "relu"))
model.add(Dense(25, activation="relu"))
model.add(Dense(1, activation="linear"))
```

```python
model.summary()
```

```python
model.compile(optimizer = "adam", loss="mean_squared_error")
```

```python
epochss = model.fit(X_train, y_train, epochs = 25, batch_size = 25, verbose = 1, validation_split=0.2)
```

```python
epochss.history.keys()
```

```python
plt.plot(epochss.history["loss"])
plt.plot(epochss.history["val_loss"])
plt.title("Model Loss During Training")
plt.legend(["Training Loss", "Validation Loss"])
```

```python
#prediction timeee
X.columns
```

```python
new_test_data = np.array([0,26,100000, -125, 50000])
y_predi = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, y_predi)
```

```python

plt.scatter(y_test, y_predi, alpha=0.5)
plt.xlabel("True Values (y_test)")
plt.ylabel("Predicted Values (y_predi)")
plt.title("True vs. Predicted Car Prices")
plt.show()
```
